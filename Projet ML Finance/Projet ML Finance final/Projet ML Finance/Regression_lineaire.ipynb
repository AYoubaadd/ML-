```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
sns.set()
```

```python
data=pd.read_csv("financials.csv")
data = data.dropna()
data.isnull().sum()
```

```python
X=data['Price']
Y=data['Price/Earnings']

```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2 ,random_state=365)
X_train_m =X_train.values.reshape(-1,1)
Y_train_m =Y_train.values.reshape(-1,1)
X_test_m =X_test.values.reshape(-1,1)
Y_test_m =Y_test.values.reshape(-1,1)

```

```python
reg=LinearRegression()
reg.fit(X_train_m,Y_train_m)

```

```python
Y_pred=reg.predict(X_test_m)
```

```python
plt.scatter(X_test,Y_test,color='red')
plt.plot(X_test,Y_pred,color='blue')
plt.title('evolotion de prix par taille')
```

```python
reg.score(X_train_m,Y_train_m)
```
